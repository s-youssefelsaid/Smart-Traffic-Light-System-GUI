```python
@app.callback(
    [Output('total','children'), Output('approved','children'), Output('rejected','children'), Output('rate','children'),
     Output('pie','figure'), Output('credit-bar','figure'), Output('scatter','figure'), Output('dti','figure'), Output('table','data')],
    [Input('status','value'), Input('gender','value'), Input('credit','value'), Input('edu','value'), Input('income-slider','value')]
)
def update(status, gender, credit, edu, income_range):
    dff = df[(df['Income'] >= income_range[0]) & (df['Income'] <= income_range[1])]
    if status: dff = dff[dff['Loan_Status'].isin(status)]
    if gender: dff = dff[dff['Gender'].isin(gender)]
    if credit: dff = dff[dff['Credit_History'].isin(credit)]
    if edu: dff = dff[dff['Education'].isin(edu)]

    total = len(dff)
    approved = len(dff[dff['Loan_Status']=='Approved'])
    rejected = len(dff[dff['Loan_Status']=='Rejected'])
    rate = f"{(approved/total*100):.1f}%" if total>0 else "0%"

    pie = px.pie(dff, names='Loan_Status', color_discrete_sequence=['#27ae60','#e74c3c'], hole=0.4)
    bar = px.histogram(dff, x='Credit_History', color='Loan_Status', barmode='group', color_discrete_sequence=['#27ae60','#e74c3c'])
    scatter = px.scatter(dff, x='Income', y='Loan_Amount', color='Loan_Status', size='Dependents', color_discrete_sequence=['#27ae60','#e74c3c'])
    dti_hist = px.histogram(dff, x='DTI', color='Loan_Status', nbins=40, color_discrete_sequence=['#27ae60','#e74c3c'])

    return f"{total:,}", f"{approved:,}", f"{rejected:,}", rate, pie, bar, scatter, dti_hist, dff.to_dict('records')
```
